```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('student_data.csv')
#Count the number of students who have an HSC GPA greater than 3.5
num_students_hsc_gpa = (df['hsc_gpa'] > 3.5).sum()
print("Number of students with HSC GPA > 3.5:", num_students_hsc_gpa)
```

```python
#There is no existence of a Chemical Engineering department in this data frame
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('student_data.csv')
# Display students from "Chemical Engineering" who have a CGPA above 3.0
engineering_students = df[(df['department'] == 'Engineering') & (df['cgpa'] > 3.0)]
print(engineering_students)

```

```python
# Calculate the average attendance for all students who have studied abroad.
# Replace multiple values
df['abroad_study'] = df['abroad_study'].replace({'true': 1})
print(df)
abroad_students = df[df['abroad_study'] == 1]
average_attendance_abroad = abroad_students['attendance'].mean()
print(f"The average attendance for all students who have studied abroad is {average_attendance_abroad}.")

```

```python
#Generate a line chart comparing the number of students from each department over the years
import matplotlib.pyplot as plt
department_counts = df.groupby(['passing_year', 'department']).size().unstack()
department_counts.plot(kind='line', figsize=(10, 6))
plt.xlabel('passing_year')
plt.ylabel('Number of Students')
plt.title('Number of Students from Each Department Over the Passing_Years ')
plt.legend(title='Department')
plt.show()
```
